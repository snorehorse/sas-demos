## Grab some data from SASHELP.CARS

```sas
data s01a_test_data;
    set sashelp.cars(keep=make model Type drivetrain Origin MSRP Cylinders);
    model = left(model);
run;

/* Create some bad data that we will detect in a later step */
proc sql;
	insert into s01a_test_data

	/* Bad cylinder counts */
	values ("Toyota", "RAV5", "SUV", "Asia", "Front", 25000, -4) 
	values ("Toyota", "RAV8a", "SUV", "Asia", "Front", 31000, 80)
	values ("Toyota", "RAV8b", "SUV", "Asia", "Front", 31000, 3.5)

	/* Bad prices */
	values ("Audi", "Foo", "SUV", "Asia", "Front", 58, 6)
	values ("Audi", "Bar", "SUV", "Asia", "Front", -25000, 6)
	values ("Audi", "Llama", "SUV", "Asia", "Front", ., 6)

	/* Duplicate record based on measure_month, make, model, drivetrain */
	values ("Saab", "9-5 Aero", "Wagon", "Europe", "Front", 50000, 4) 

	/* Bad Type */
	values ("Nissan", "Who Cares", "BadValue", "USA", "Front", 40000, 4) 

	/* Bad Drive Train */
	values ("Nissan", "Awesome", "Wagon", "USA", "Apple", 45000, 4) 

	/* MSRP outlier */
	values ("Ford", "Loaf", "Wagon", "USA", "Front", 987654321, 4) 
	;
quit;

proc print; run;
```
